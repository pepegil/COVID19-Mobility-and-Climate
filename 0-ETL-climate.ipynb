## Transforming precipitation data

**Diferent formats are present : numbers ,strings, decimal separation,'nan, 'Ip' value.**

```python
clima.prec.unique()
```

```python
 # all strings
clima['prec'] = clima['prec'].astype(str)
```

```python
def formato_prec(row): ## Translates strings into floats representing liters
    if row.prec== 'nan' :
        precN = 0.0
    elif row.prec== 'Ip':
        precN = 0.0
    else :
        litros = float(row.prec[:-2])
        decima = float(row.prec[-1:])
        precN = (litros) + (decima)/10.0
    
    return precN 
```

```python
prec = clima.apply(lambda row: formato_prec(row), axis=1).rename('prec')
```

```python
col_eliminar_met = ['prec']
clima.drop(columns = col_eliminar_met, axis = 1, inplace = True) 
clima.insert(4,'prec',prec)
```

# Feature Engineering

## Absolute Humidity

\begin{equation}
  ha = \frac{P}{RvT}
\end{equation}

*[ha] = K/m3,  [P] = Pa, [T] = ºK, Rv = 461 [J/K ºK]*

```python
def humedad_absoluta(row): 
    P = (row.presMax - row.presMin) / 2.0
    ha = 1000000.0 * (P/100.0) / (461 * (row.tmed + 273.15)) # Kg/liter
          
    return ha
```

```python
ha = datos_clima.apply(lambda row: humedad_absoluta(row), axis=1) #.rename('provincia')
datos_clima.insert(10,'ha', ha)# evolucion.head()
datos_clima.info()
```

```python
datos_clima.ha.describe()
```

## Saturation presion

\begin{equation}
  Ps(T) = 6.11 e^\frac{7.5T}{237.3 + T}
\end{equation}

     # T  ºC,  ps [hPa=mb=100Pa]

```python
def presion_saturacion (row) : # Clausius-Clapeyron, partial pressure of saturated air 
    # T  ºC,  ps [hPa=mb=100Pa]
    ps = 6.11 * (np.exp( (7.5*row.tmed) / (237.3 + row.tmed) ) ) 
    
    return ps
```

```python
psat = datos_clima.apply(lambda row: presion_saturacion(row), axis=1) #.rename('provincia')
datos_clima.insert(11,'psat', psat)# evolucion.head()
datos_clima.info()
```

```python
datos_clima.psat.describe()
```

## Air volume and P-T variation

```python
# En pascales , grados Kelvin  y litros
datos_clima["vol_min"] = 1000.0 * datos_clima["tmed"].apply(
    lambda datos_clima: datos_clima + 273) / datos_clima["presMax"].apply(lambda datos_clima: datos_clima*100)

datos_clima["vol_max"] = 1000.0 * datos_clima["tmed"].apply(
    lambda datos_clima: datos_clima + 273) / datos_clima["presMin"].apply(lambda datos_clima: datos_clima*100)

datos_clima["delta_V"] = datos_clima["vol_max"] - datos_clima["vol_min"]
datos_clima["delta_P"] = datos_clima["presMax"] - datos_clima["presMin"]
datos_clima["delta_T"] = datos_clima["tmax"] - datos_clima["tmin"]
```
